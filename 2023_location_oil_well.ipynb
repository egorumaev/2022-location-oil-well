Оценим, какие признаки дают в каждом из датасетов наибольшую информацию о целевом признаке. Это расширит наше понимание о важности каждого из признаков

```python
# разделим каждый из датасетов на features_n (признаки, без учета 'id')
# и target_n (целевой признак)

features_1 = reg1[['f0', 'f1', 'f2']]
target_1 = reg1['product']
features_1.shape, target_1.shape
```

```python
features_2 = reg2[['f0', 'f1', 'f2']]
target_2 = reg2['product']
features_2.shape, target_2.shape
```

```python
features_3 = reg3[['f0', 'f1', 'f2']]
target_3 = reg3['product']
features_3.shape, target_3.shape
```

Определим для каждого из датасетов, каике признаки дают большую информацию о целевом признаке

```python
mi_1 = mutual_info_regression(features_1, target_1, random_state = 1)

plt.figure(figsize = (8, 3))
ax = sns.barplot(x = features_1.columns.tolist(), y = mi_1)
plt.xticks(rotation = 70)
plt.title('Рис. 7 - Mutual Information между целевым признаком и признаками (reg1)', \
          fontsize = 12)

for bar in ax.patches:
    if bar.get_height() > 0.04:
        bar.set_color('indianred')
    else:
        bar.set_color('cadetblue')

plt.show()
```

```python
mi_2 = mutual_info_regression(features_2, target_2, random_state = 1)

plt.figure(figsize = (8, 3))
ax = sns.barplot(x = features_2.columns.tolist(), y = mi_2)
plt.xticks(rotation = 70)
plt.title('Рис. 8 - Mutual Information между целевым признаком и признаками (reg2)', \
          fontsize = 12)

for bar in ax.patches:
    if bar.get_height() > 0.04:
        bar.set_color('indianred')
    else:
        bar.set_color('cadetblue')

plt.show()
```

```python
mi_3 = mutual_info_regression(features_3, target_3, random_state = 1)

plt.figure(figsize = (8, 3))
ax = sns.barplot(x = features_3.columns.tolist(), y = mi_3)
plt.xticks(rotation = 70)
plt.title('Рис. 9 - Mutual Information между целевым признаком и признаками (reg3)', \
          fontsize = 12)

for bar in ax.patches:
    if bar.get_height() > 0.04:
        bar.set_color('indianred')
    else:
        bar.set_color('cadetblue')

plt.show()
```

Результаты по регионам reg1 и reg3 сопоставимы. Регион reg2 резко отличается в сравнении с ними: признаки 'f2' 'f0' дают значительно больше информации о целевом признаке. Согласно выставленного задания каждая из трех выборок должна быть разделена на обучающую и валидационную выборки в соотношении 75:25

На первом шаге разделения датасетов на выборки выделим в каждом датасете признаки и целевой признак

Целевым признаком в каждом из трех датасетов является столбец 'product'

Признаки представлены тремя столбцами: f0, f1, f2

Столбец 'id' не будет включен в в состав признаков - он никак не связан с целевым признаком

```python
# создадим функцию для разделения выборок с признаками и целевым признаком
# по каждому региону на обучающую и валидационную
def datasets_split(features, target):
    features_train, features_valid, \
    target_train, target_valid = train_test_split(
    features,
    target,
    test_size = 0.25,
    random_state = 1)
    return features_train, features_valid, target_train, target_valid
```

### **4.1 Расчет прибыли по отобранным скважинам и предсказаниям модели**

В рамках раздела последовательно выполнены следующие действия:

**(1)** выбраны скважины с максимальными значениями предсказаний, полученных во 2-м разделе

**(2)** просуммировано целевое значение объёма сырья, соответствующее полученным предсказаниям

**(3)** рассчитана прибыль для полученного объёма сырья

```python
# напишем функцию для расчета прогнозируемой прибыли в руб.
def gain(target, predictions, count, price, budget):
    predictions_sorted = predictions.sort_values(ascending = False)
    selected = target[predictions_sorted.index][:count]
    total_income = selected.sum() * price
    gain = total_income - budget
    return gain
```

```python
# вызовем функцию расчета прогнозируемой прибыли для региона 1
reg1_predicted_income = gain(
    target_valid_1,
    predict_1,
    POINTS_BEST,
    UNIT_PRODUCT_COST,
    BUDGET)
```

### **5.1 Подготовка к расчету рисков и распределения прибыли для каждого из трех исследуемых регионов**

---

Для расчета рисков и распределения прибыли применим технику Bootstrap, найдем среднюю прибыль, 95%-й доверительный интервал и риск убытков. Убыток рассматриваем в расчетах как отрицательную прибыль. На каждой итерации Bootstrap будет исследована выборка из 500 точек (скважин)

Для расчетов уровня рисков и распределения прибыли с помощью техники Bootstrap будет использована функция gain(target, predictions, count, price, budget) из п. 4.1. На ее основе будет создана новая функция, которая кроме расчета прибыли по региону на каждой итерации позволит рассчитать среднее значение прибыли по региону по 1000 выборкам, границы 95%-го межквартильного интервала и уровень риска

```python
def risks_and_average_profit(target, predictions, iterations, \
                             points_exploration, count, price, budget):
    state = np.random.RandomState(1)
    iterations_items = []
    for i in range(iterations):
        target_subsample = target.sample(n = points_exploration,
                                         replace = True,
                                         random_state = state)

        predictions_subsample = predictions[target_subsample.index]

        iterations_items.append(gain(target_subsample, \
                                     predictions_subsample, \
                                     POINTS_BEST, \
                                     UNIT_PRODUCT_COST, \
                                     BUDGET))

    # изменяем тип массива numpy arrey, полученного с помощью Bootstraping,
    # на Series
    iterations_items = pd.Series(iterations_items)

    # вычислим среднее значение прибыли
    # по 1000 выборкам двухсот лучших скважин
    mean_profit = iterations_items.mean()

    # для определения нижней границы 95%-го доверительного интервала
    # рассчитаем 2,5% квантиль
    lower = round(iterations_items.quantile(0.025), 2)

    # для определения верхней границы 95%-го доверительного интервала
    # рассчитаем 97,5% квантиль
    upper = round(iterations_items.quantile(0.975), 2)

    # для удобства работы с возвращенными функцией значениями
    # создадим объект quantile_interval, включающим информацию
    # о нижней и верхней границе 95%-го доверительного интервала
    quantile_interval = (lower, upper)

    # для расчета вероятности получения отрицательной прибыли (убытка)
    # посчитаем количество отрицательных результатов в 1000 выборках,
    # полученных с помощью Bootstrap, и разделим на количество выборок (1000)
    negative_profit = iterations_items.apply(lambda x: 1 if x < 0 else 0)
    risk = (negative_profit.sum() / 1000) * 100

    return mean_profit, risk, quantile_interval
```
